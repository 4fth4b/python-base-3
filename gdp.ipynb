```python
def IQR_values(df1,Column_name):
    q1=df1[Column_name].quantile(0.25)
    q2=df1[Column_name].quantile(0.50)
    q3=df1[Column_name].quantile(0.75)
    IQR=q3-q1
    low_limit=q1-1.5*IQR
    upper_limit=q3+1.5*IQR
    return low_limit,upper_limit
```

```python
df1
```

```python
l=['Population in thousands (2017)','GDP per capita (current US$)']
j={}
for i in l:
   low_limit,upper_limit= IQR_values(df1,i)
   j[i]=[low_limit,upper_limit]
    
```

```python
j
```

```python
plt.boxplot(df1['GDP per capita (current US$)'])
```

```python
x=[]
for i in l:
    a=df1[(df1[i]>j[i][1])|(df1[i]<j[i][0])].index.values
    x.append(a)
```

```python
x
```

```python
for i in range(len(x)):
    df1.loc[x[i],l[i]]=df1[l[i]].mean()
```
